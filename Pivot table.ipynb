```python
import pandas as pd
import numpy as np
from datetime import datetime
from datetime import timedelta 
import time
```

```python
df = pd.read_csv("./data/marathon_results_2016.csv",usecols=['Bib','40K','Half','Pace','Age','M/F','Country','State','City'],index_col="Bib")
```

```python
df.head()
```

```python
df['40K'] = df['40K'].apply(pd.to_timedelta)
df['Half'] = df['Half'].apply(pd.to_timedelta)
df['TotalSeconds'] = df['40K'].apply(lambda x: timedelta.total_seconds(x))
df['HalfSeconds'] = df['Half'].apply(lambda x: timedelta.total_seconds(x))
```

```python
df.head()
```

```python
df.pivot_table(index="Age",columns="M/F",aggfunc="mean",values="TotalSeconds").head()
```

```python
df.pivot_table(index="Age",columns="M/F",aggfunc="mean",values=["TotalSeconds","HalfSeconds"]).swaplevel(i=0,j=1,axis="columns").head()
```

```python
df.pivot_table(index="Age",columns="M/F",aggfunc="mean",values=["TotalSeconds","HalfSeconds"]).swaplevel(i=0,j=1,axis="columns").sort_index(axis="columns").head()
```

```python
df.pivot_table(index="Age",columns="M/F",aggfunc="count",values=['City']).fillna(0).astype(int)
```

```python
df.pivot_table(index="Country",aggfunc='count',values='City').head()
```

```python
df.pivot_table(index="Country",aggfunc='count',values='City').sort_values(by="City",ascending=False).head()
```

```python
df.pivot_table(index=['Country','State','City'],values='TotalSeconds',aggfunc='mean').head()
```

```python
df.pivot_table(index=['Country','State','City'],values='TotalSeconds',aggfunc='mean').sort_values("TotalSeconds",ascending=False).head()
```
